# Number of files

```python

allFolders = os.listdir()

fileCount = 0
for each in allFolders:
    if "." in each:
        continue
    folders1 = os.listdir(each)
    for each1 in folders1:
        folders2 = os.listdir(each + "/" + each1)
        for each2 in folders2:
            if ".csv" in each2:
                fileCount += 1
            else:
                print("file not counted", each2)
        
        
print("Total number of files in this dataset: ",fileCount)

# format
# total number of files in this dataset
```

# Unique File Names

```python

allFolders = os.listdir()

allFileNames = {}
for each in allFolders:
    if "." in each:
        continue
    folders1 = os.listdir(each)
    for each1 in folders1:
        folders2 = os.listdir(each + "/" + each1)
        for each2 in folders2:
            thisFileName = each2.split("_")[:-2]
            thisFileName = "_".join(thisFileName)
            if thisFileName not in allFileNames.keys():
                allFileNames[thisFileName] = 1
            else:
                allFileNames[thisFileName] += 1
                
        
    
for each in allFileNames:
    print(each + ": " + str(allFileNames[each]))

# format
# file name : total number of files with that name (the date and time may be differnet)
```

# Unique file names generalized

```python
# the above file naems have numebrs in them. If we remove the numbers we can group them.

filteredFileNames = {}

for each in allFileNames:
    result = ''.join([i for i in each if not i.isdigit()])
    if result in filteredFileNames.keys():
        filteredFileNames[result] += allFileNames[each]
    else:
        filteredFileNames[result] = allFileNames[each]
        
for each in filteredFileNames:
    print(each + ": " + str(filteredFileNames[each]))
    
# format
# file name : number of files of that type
```

# Collect all the column names from different files. For every name count how many files have this column

```python
allFolders = os.listdir()

allColumns = {}
for each in allFolders:
    if "." in each:
        continue
    folders1 = os.listdir(each)
    for each1 in folders1:
        folders2 = os.listdir(each + "/" + each1)
        for each2 in folders2:
            if ".csv" not in each2:
                continue
            with open(each + "/" + each1 + "/" + each2) as f:
                firstLine = f.readline()
            allCols = firstLine.split(",")
            allColsUnique = set(allCols)
            if len(allCols) != len(allColsUnique):
                print("There are duplicate columns in ",each + "/" + each1 + "/" + each2)
            for eachCol in allCols:
                if eachCol not in allColumns.keys():
                    allColumns[eachCol] = 1
                else:
                    allColumns[eachCol] += 1

                    
# there are lots of duplicated columns in this dataset
# this suggests that this dataset has already been tokenized 
```

```python
allColumns=dict(sorted(allColumns.items(), key=lambda item: item[1], reverse=True))
print(json.dumps(allColumns, indent=4))

# Format
# Column name : how many files have this column
```

- create a table with properties of all files in the data set (name, size, number of columns and rows, etc)
- think about similarly measure on the files to be used for clustering

```python
def getFileSize(path):
    command = "du -sh " + path
    output = subprocess.check_output(command, shell=True, encoding='utf-8')
    return output.split("\t")[0]

def getNumberOfLines(path):
    command = "wc -l " + path
    output = subprocess.check_output(command, shell=True, encoding='utf-8')
    return output.split()[0]

allFolders = os.listdir()
fileNames = []
for each in allFolders:
    if "." in each:
        continue
    folders1 = os.listdir(each)
    for each1 in folders1:
        folders2 = os.listdir(each + "/" + each1)
        for each2 in folders2:
            if ".csv" in each2:
                fileNames.append(each + "/" + each1 + "/" + each2)
            else:
                print("file not counted", each2)

fileDetails = []
for eachFile in fileNames:
    thisFileDetails = {
        'name' : eachFile
    }
    with open(eachFile) as f:
        cols = f.readline().split(",")
    thisFileDetails['columns'] = len(cols)
    thisFileDetails['size'] = getFileSize(eachFile)
    thisFileDetails['line_count'] = getNumberOfLines(eachFile)
    fileDetails.append(thisFileDetails)        
```

```python
with open("file_info.txt",'w') as f:
    f.write(json.dumps(fileDetails, indent=4))
```
